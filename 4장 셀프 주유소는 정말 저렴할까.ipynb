```python
import pandas as pd
from glob import glob
```

```python
#현재 파일 25개=>1개씩 읽어서 데이터프레임으로 변환=>1개씩 읽어서 데이터프레임으로 변환
# 25개 파일명이 들어 있는 파일 목록 생성
# 파일 목록 생성 : glob()
glob('./data_01/지역*.xls')
```

```python
station_files = glob('./data_01/지역*')
```

```python
type(station_files)
```

```python
# 25개 구를 반복해서 한개씩 읽어서 데이터프레임으로 만든다.
# 생성된 데이터 프레임을 리스트에 저장.
# concat에 리스트를 넣으면 하나로 저장된다.

tmp_raw =[]
    
for file_name in station_files:
    tmp = pd.read_excel(file_name, header=2) #header를 2로 줘서 첫 2개행 삭제
    tmp_raw.append(tmp)
tmp_raw
```

```python
tmp_raw[0]
```

```python
len(tmp_raw)
```

```python
station_raw = pd.concat(tmp_raw)
```

```python
station_raw.head()
```

```python
station_raw.info()
```

```python
stations = pd.DataFrame(
    {
        '상호':station_raw['상호']
        ,'주소':station_raw['주소']
        ,'가격':station_raw['휘발유']
        ,'셀프여부':station_raw['셀프여부']
        ,'상표':station_raw['상표']
    }
)
stations.head()
```

```python
# 구 컬럼 생성. : 주소에서 구를 추출. => 리스트
stations['구'] = [addr.split()[1] for addr in stations['주소']]#내포된 for문은 매우 중요하다.
```

```python
stations['구'].unique()
```

```python
# 가격 입력 X =>'-'
stations[stations['가격']=='-']
```

```python
# '-'이 아닌 주유소 추출
stations = stations[stations['가격']!='-']
```
